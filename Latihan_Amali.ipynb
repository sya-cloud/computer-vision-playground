```python
# Import pustaka
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Menjana data rawak
np.random.seed(42)  # Supaya hasil boleh diulang
data = np.random.randint(1, 101, size=(10, 3))

# Membina DataFrame
df = pd.DataFrame(data, columns=["A", "B", "C"])

# Paparkan DataFrame
print(df)

# Lukis graf
plt.figure(figsize=(8,5))
plt.plot(df["A"], label="A", marker="o")
plt.plot(df["B"], label="B", marker="o")
plt.plot(df["C"], label="C", marker="o")

# Hiasan graf
plt.title("Graf Data Rawak")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.legend()
plt.grid(True)

# Paparkan graf
plt.show()
```

```python
# Import pustaka yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Menetapkan benih nombor rawak untuk hasil yang konsisten
np.random.seed(42)

# Menjana data rawak bersaiz 10 x 3
data = np.random.randint(1, 101, size=(10, 3))

# Membina DataFrame menggunakan data yang dijana
df = pd.DataFrame(data, columns=["A", "B", "C"])

# Memaparkan DataFrame
print(df)

# Menetapkan saiz graf
plt.figure(figsize=(8, 5))

# Melukis graf bagi setiap lajur
plt.plot(df["A"], marker="o", label="A")
plt.plot(df["B"], marker="o", label="B")
plt.plot(df["C"], marker="o", label="C")

# Menambah tajuk dan label paksi
plt.title("Graf Data Rawak")
plt.xlabel("Indeks")
plt.ylabel("Nilai")

# Memaparkan legenda dan grid
plt.legend()
plt.grid(True)

# Memaparkan graf
plt.show()
```
